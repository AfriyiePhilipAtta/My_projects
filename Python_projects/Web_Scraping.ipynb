```python
#In this lab Iused APIs and BeautifulSoup to extract average height and weight chart data from https://schoolsnews.org/average-height-and-weight-chart-by-age


from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://schoolsnews.org/average-height-and-weight-chart-by-age'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
soup
```

```python
table = soup.find('table', class_ = 'dataframe')
table
```

```python
columns = table.find_all('th')
columns
```

```python
column_list = [th.text for th in columns]
column_list
```

```python
df = pd.DataFrame(columns = column_list)
df
```

```python
data = soup.find_all('tr')
for row in data[1:20]:
    row_data = row.find_all('td')
    individual_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_data
df    
```
